#### There are zero values for features that it is considered to be very unlikely. 
For  Features: “BloodPressure”, “BMI” and “Glucose”, will remove the rows where they are zero. 

```python
diabetes_dataset_modified = diabetes_dataset[(diabetes_dataset.BloodPressure != 0) & (diabetes_dataset.BMI != 0) & (diabetes_dataset.Glucose != 0)]
print(diabetes_dataset_modified.shape)
```

## 3-Training 


#### - Seperate the features and the Outcome

```python
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = diabetes_dataset_modified[features]
y = diabetes_dataset_modified.Outcome
```

#### - Split the data into training and testing subsets


```python

# Train/Test split
#Store the results in X_train, X_test, y_train, and y_test
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = diabetes_dataset_modified.Outcome, random_state=0)

# Show the results of the split
print "Training set has {} samples.".format(X_train.shape[0])
print "Testing set has {} samples.".format(X_test.shape[0])
```

## Model Selection


### Fit a classifier to the training data

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import StratifiedKFold
from sklearn.metrics import accuracy_score
```

```python
models = []

models.append(('SVC', SVC(gamma=0.001))) 
models.append(('SVC', SVC(kernel='rbf')))

models.append(('KNN', KNeighborsClassifier()))
models.append(('DT', DecisionTreeClassifier()))
models.append(('LR', LogisticRegression(solver='lbfgs', max_iter=4000)))
```

#### Train and predict using a classifier based on accuracy score

```python
classifiersNames = []
accuracyScores = []

for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracyScores.append(accuracy_score(y_test, y_pred))
    classifiersNames.append(name)

tr_split = pd.DataFrame({'Classifier Name ': classifiersNames, 'Score': accuracyScores})
print(tr_split)
```

## 4- Using K-Fold cross validation


```python
strati_k_fold = StratifiedKFold(y, n_folds=10, random_state=10)

classifiersNames = []
accuracyScores = []

for name, model in models:
    
    score = cross_val_score(model, X, y, cv=strati_k_fold, scoring='accuracy').mean()
    classifiersNames.append(name)
    accuracyScores.append(score)

k_fold_cross_value = pd.DataFrame({'Classifier Name ': classifiersNames, 'Score': accuracyScores})
print(k_fold_cross_value)
```

#### plot the accuracy scores

```python
axis = sns.barplot(x = 'Classifier Name ', y = 'Score', data = k_fold_cross_value)
axis.set(xlabel='Classifier', ylabel='Accuracy')

for p in axis.patches:
    height = p.get_height()
    axis.text(p.get_x() + p.get_width()/2, height + 0.005, '{:1.4f}'.format(height), ha="center") 
    
plt.show()
```

## 5 - Feature Selection

### DecisionTreeClassifier

```python
from sklearn.feature_selection import RFECV
```

```python
dt_model =  DecisionTreeClassifier()

dt_rfecv = RFECV(estimator=dt_model, step=1, cv=strati_k_fold, scoring='accuracy')
dt_rfecv.fit(X, y)

plt.figure()
plt.title('DecisionTreeClassifier score vs No of Features')
plt.xlabel("Number of selected features ")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(dt_rfecv.grid_scores_) + 1), dt_rfecv.grid_scores_)
plt.show()
```

```python
feature_importance = list(zip(features, dt_rfecv.support_))

important_new_features = []

for key,value in enumerate(feature_importance):
    if(value[1]) == True:
        important_new_features.append(value[0])
        
print(important_new_features)
```

#### Find the accuracy with important features

```python
X_new_dt = diabetes_dataset_modified[important_new_features]

initial_score = cross_val_score(dt_model, X, y, cv=strati_k_fold, scoring='accuracy').mean()
print("Initial accuracy : {} ".format(initial_score))

fe_score = cross_val_score(dt_model, X_new_dt, y, cv=strati_k_fold, scoring='accuracy').mean()
print("Accuracy after Feature Selection : {} ".format(fe_score))
```
